```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
print(iris)
```

```python
correlation = iris.corr()
```

```python
sns.heatmap(iris.corr(), cmap='coolwarm',annot=True) #annot false dile majkhaner number gula ashbe na
```

```python
sns.set(style="whitegrid", color_codes=True)
```

```python
sns.lmplot(x="sepal_length", y="petal_length", hue="species", data=iris, palette="Set1", markers=["o", "s", "D"])
```

```python
sns.lmplot(x="sepal_length", y="petal_length", hue="species",data=iris, palette="Set1",fit_reg=False, markers=["o","s","D"])
```

### pairplot

```python
sns.pairplot(iris, size=5, vars=["sepal_length", "petal_length"],hue="species", markers=["o","s","D"])
```

```python
sns.pairplot(iris, hue="species", markers=["o","s","D"])
```

### pairplot with best fit line

```python
sns.pairplot(iris, hue="species", kind="reg", markers=["o","s","D"])
```

### boxplot

```python
sns.boxplot(data=iris,orient="h")
```

### plotting with feature ratio

```python
iris["ID"] = iris.index
iris["ratio"] = iris["sepal_length"]/iris["sepal_width"]
sns.lmplot(x="ID", y="ratio", data=iris, hue="species", markers=["o","s","D"], fit_reg=False, legend=True)
```

### strip plot

```python
sns.stripplot(x="species", y="sepal_length", data=iris)
```

```python
sns.stripplot(x = "species", y = "petal_length", data = iris, jitter = True)
```

### swarm plot

```python
sns.swarmplot(data=iris, x="species", y="petal_length")
```
